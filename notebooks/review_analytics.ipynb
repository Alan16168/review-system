# Review System 数据分析

## 目标
分析用户复盘数据，为曼哈顿计划提供数据支持

---

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

## 1. 数据加载

从 Cloudflare D1 数据库加载数据

```python
# 示例数据结构
data = {
    'user_id': [1, 2, 3, 4, 5],
    'reviews_count': [15, 8, 23, 12, 5],
    'subscription_tier': ['premium', 'free', 'premium', 'free', 'free'],
    'monthly_revenue': [19.99, 0, 19.99, 0, 0],
    'created_at': pd.date_range(start='2024-01-01', periods=5, freq='M')
}

df = pd.DataFrame(data)
df.head()
```

## 2. 用户增长分析

```python
# 付费转化率
paid_users = df[df['subscription_tier'] == 'premium'].shape[0]
total_users = df.shape[0]
conversion_rate = (paid_users / total_users) * 100

print(f"总用户数: {total_users}")
print(f"付费用户数: {paid_users}")
print(f"付费转化率: {conversion_rate:.2f}%")
```

## 3. 收入预测（曼哈顿计划）

```python
# 基于当前数据预测月收入
monthly_revenue = df['monthly_revenue'].sum()
print(f"当前月收入: ${monthly_revenue:.2f}")

# 曼哈顿计划目标
target_revenue = 37477
current_achievement = (monthly_revenue / target_revenue) * 100
print(f"\n曼哈顿计划目标: ${target_revenue}")
print(f"当前完成度: {current_achievement:.2f}%")
print(f"距离目标还差: ${target_revenue - monthly_revenue:.2f}")
```

## 4. 可视化分析

```python
# 用户分布饼图
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 订阅层级分布
tier_counts = df['subscription_tier'].value_counts()
axes[0].pie(tier_counts.values, labels=tier_counts.index, autopct='%1.1f%%')
axes[0].set_title('订阅层级分布')

# 收入构成
revenue_by_tier = df.groupby('subscription_tier')['monthly_revenue'].sum()
axes[1].bar(revenue_by_tier.index, revenue_by_tier.values)
axes[1].set_title('各层级月收入')
axes[1].set_ylabel('收入 (USD)')

plt.tight_layout()
plt.show()
```

## 5. 曼哈顿计划实施建议

```python
# 计算需要多少付费用户才能达到目标
premium_price = 19.99
needed_premium_users = target_revenue / premium_price

print(f"若全部为Premium用户($19.99/月)，需要: {needed_premium_users:.0f}个")
print(f"当前Premium用户: {paid_users}个")
print(f"还需增加: {needed_premium_users - paid_users:.0f}个Premium用户")

# 混合收入模型
print("\n--- 混合收入模型（曼哈顿计划） ---")
print("订阅收入: $11,977/月 (200 Premium + 80 Team + 20 Enterprise)")
print("信用点销售: $15,000/月 (500用户 × $30平均)")
print("模板销售: $10,500/月 (700用户 × $15平均)")
print("总计: $37,477/月")
```
