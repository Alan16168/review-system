# AI 服务成本与收益分析

## 目标
分析曼哈顿计划中 AI 写作和 AI 视频生成的成本结构

---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

## 1. AI 写作成本分析

```python
# Gemini 2.5 Pro 定价（示例）
gemini_input_cost = 0.00125  # per 1K tokens
gemini_output_cost = 0.005   # per 1K tokens

# 典型请求：3000字文章
avg_prompt_tokens = 500
avg_output_tokens = 4000  # 3000字约4000 tokens

cost_per_request = (
    (avg_prompt_tokens / 1000) * gemini_input_cost +
    (avg_output_tokens / 1000) * gemini_output_cost
)

print(f"AI写作单次成本: ${cost_per_request:.4f}")
print(f"用户支付（10 credits = $1）: $1.00")
print(f"毛利: ${1.00 - cost_per_request:.4f}")
print(f"毛利率: {((1.00 - cost_per_request) / 1.00 * 100):.2f}%")
```

## 2. AI 视频成本分析

```python
# D-ID 定价（示例）
did_cost_per_video = 1.50  # per video (5分钟)

# 用户支付
user_payment = 2.99  # $2.99/video 或 300 credits

video_margin = user_payment - did_cost_per_video
video_margin_rate = (video_margin / user_payment) * 100

print(f"AI视频单次成本: ${did_cost_per_video:.2f}")
print(f"用户支付: ${user_payment:.2f}")
print(f"毛利: ${video_margin:.2f}")
print(f"毛利率: {video_margin_rate:.2f}%")
```

## 3. 月度成本预测

```python
# 假设每月使用量（基于1000用户）
monthly_writing_requests = 2000
monthly_video_requests = 100

# 总成本
monthly_writing_cost = monthly_writing_requests * cost_per_request
monthly_video_cost = monthly_video_requests * did_cost_per_video
total_ai_cost = monthly_writing_cost + monthly_video_cost

# 总收入（从曼哈顿计划）
monthly_ai_revenue = 15000  # 信用点销售

# 净利润
net_profit = monthly_ai_revenue - total_ai_cost
profit_margin = (net_profit / monthly_ai_revenue) * 100

print("=== 月度 AI 服务成本分析 ===")
print(f"\nAI写作成本: ${monthly_writing_cost:.2f}")
print(f"AI视频成本: ${monthly_video_cost:.2f}")
print(f"AI总成本: ${total_ai_cost:.2f}")
print(f"\nAI服务收入: ${monthly_ai_revenue:.2f}")
print(f"净利润: ${net_profit:.2f}")
print(f"利润率: {profit_margin:.2f}%")
```

## 4. 盈亏平衡分析

```python
# 固定成本（每月）
fixed_costs = {
    '服务器/Cloudflare': 50,
    '数据库存储': 20,
    '其他工具': 30
}

total_fixed_cost = sum(fixed_costs.values())

# 盈亏平衡点
breakeven_revenue = total_fixed_cost + total_ai_cost
breakeven_users = breakeven_revenue / 30  # 假设每用户$30平均消费

print("=== 盈亏平衡分析 ===")
print(f"\n月固定成本: ${total_fixed_cost:.2f}")
print(f"月变动成本（AI）: ${total_ai_cost:.2f}")
print(f"盈亏平衡收入: ${breakeven_revenue:.2f}")
print(f"盈亏平衡用户数: {breakeven_users:.0f}人")
print(f"\n当前目标（曼哈顿计划）:")
print(f"  - 用户数: 1000人")
print(f"  - 月收入: $37,477")
print(f"  - 远超盈亏平衡点 ✅")
```

## 5. 成本优化建议

```python
print("=== 成本优化策略 ===")
print("\n1. **批量折扣**:")
print("   - 与API提供商协商批量折扣（>10K请求/月）")
print("   - 预期节省: 20-30%")

print("\n2. **缓存机制**:")
print("   - 缓存常见问题的AI回答")
print("   - 预期节省API调用: 15-20%")

print("\n3. **Token优化**:")
print("   - 优化prompt，减少输入token数")
print("   - 预期节省: 10-15%")

print("\n4. **订阅模式**:")
print("   - 鼓励用户订阅而非按次付费")
print("   - 降低单次交易成本")
print("   - 提高用户LTV")

# 优化后的成本预测
optimized_cost = total_ai_cost * 0.65  # 35%节省
optimized_profit = monthly_ai_revenue - optimized_cost

print(f"\n优化后AI成本: ${optimized_cost:.2f}")
print(f"优化后净利润: ${optimized_profit:.2f}")
print(f"利润提升: ${optimized_profit - net_profit:.2f}")
```

## 6. 可视化

```python
# 成本构成饼图
costs = {
    'AI写作': monthly_writing_cost,
    'AI视频': monthly_video_cost,
    '固定成本': total_fixed_cost
}

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 成本构成
axes[0].pie(costs.values(), labels=costs.keys(), autopct='%1.1f%%')
axes[0].set_title('月度成本构成')

# 收入 vs 成本对比
comparison = ['收入', '成本', '净利润']
values = [monthly_ai_revenue, total_ai_cost + total_fixed_cost, net_profit - total_fixed_cost]
colors = ['green', 'red', 'blue']

axes[1].bar(comparison, values, color=colors)
axes[1].set_title('收入 vs 成本')
axes[1].set_ylabel('金额 (USD)')

plt.tight_layout()
plt.show()
```
